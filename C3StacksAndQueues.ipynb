```python
class StackByList:
    def __init__(self):
        self.values = []
        self.size = 0
        
    def push(self, value):
        self.values.append(value)
        self.size += 1
        
    def pop(self):
        if (self.size == 0):
            print('this stack is empty 1')
            return
        
        self.size -= 1
        return self.values.pop()
    
    def peek(self):
        if (self.size == 0):
            print('this stack is empty 2')
            return
        
        return self.values[self.size-1]

    def printList(self):
        print(self.values)
```

## 3.1 Three in One 2

```python
class MultiStack:      
    def __init__(self, numberOfStacks, defaultSize):
        self.numberOfStacks = numberOfStacks
        self.values = [0] * numberOfStacks * defaultSize
        self.start = []
        self.capacity = []
        self.size = []
        for i in range(numberOfStacks):
            self.start.append(defaultSize * i)
            self.capacity.append(defaultSize)
            self.size.append(0)
    
    def push(self, stackNum, value):
        if (self.allStacksAreFull()):
            print('all stacks are full')
            return None
        
        if (self.isFull(stackNum)):
            self.expand(stackNum)
            
        self.size[stackNum] += 1
        self.values[self.lastElementIndex(stackNum)] = value
    
    def pop(self, stackNum):
        if (self.isEmpty(stackNum)):
            print('this stack is empty')
            return None
        
        value = self.values[self.lastElementIndex(stackNum)]
        self.values[self.lastElementIndex(stackNum)] = 0
        self.size[stackNum] -= 1
        return value
    
    def peek(self, stackNum):
        return self.values[self.lastElementIndex(stackNum)]
    
    def shift(self, stackNum):
        print('   Shifting ', stackNum)
        
        if (self.size[stackNum] >= self.capacity[stackNum]):
            self.shift((stackNum + 1) % self.numberOfStacks)
            self.capacity[stackNum] += 1
        
        index = self.lastCapacityIndex(stackNum)
        while (self.isWithinStackCapacity(stackNum, index)):
            self.values[index] = self.values[self.previousIndex(index)]
            index = self.previousIndex(index)
        
        self.values[self.start[stackNum]] = 0
        self.start[stackNum] = self.nextIndex(self.start[stackNum])
        self.capacity[stackNum] -= 1
        
    def expand(self, stackNum):
        self.shift((stackNum + 1) % self.numberOfStacks)
        self.capacity[stackNum] += 1
    
    def isWithinStackCapacity(self, stackNum, index):
        if (index < 0 or index >= len(self.values)):
            return False

        contiguousIndex = index + len(self.values) if (index < self.start[stackNum]) else index
        end = self.start[stackNum] + self.capacity[stackNum]
        return self.start[stackNum] <= contiguousIndex and contiguousIndex < end

    def lastCapacityIndex(self, stackNum):
        return self.adjustIndex(self.start[stackNum] + self.capacity[stackNum] - 1)

    def lastElementIndex(self, stackNum):
        return self.adjustIndex(self.start[stackNum] + self.size[stackNum] - 1)

    def isFull(self, stackNum):
        return self.size[stackNum] == self.capacity[stackNum]

    def isEmpty(self, stackNum):
        return self.size[stackNum] == 0
    
    def allStacksAreFull(self):
        return sum(self.size) == len(self.values)
    
    def adjustIndex(self, index):
        max = len(self.values)
        return ((index % max) + max) % max
    
    def nextIndex(self, index):
        return self.adjustIndex(index + 1)
    
    def previousIndex(self, index):
        return self.adjustIndex(index - 1)
    
    def printList(self):
        print(self.values)
    
'''
a = MultiStack(3, 5)
a.printList()

a.push(2, 5)
a.push(2, 6)
a.push(2, 7)
a.push(2, 8)
a.push(2, 9)
a.printList()

a.push(2, 10)
a.printList()

a.push(2, 11)
a.printList()

print(a.pop(2))
a.printList()

a.push(2, 11)
a.printList()

a.push(0,1)
a.printList()

a.push(0,2)
a.push(0,3)
a.push(0,4)
a.push(0,5)
a.push(0,6)
a.push(0,7)
a.printList()

a.push(1,-1)
a.printList()

a.pop(1)
a.printList()
'''
```

## 3.2 Stack Min

## 3.3 Stack of Plates

```python
class SetOfStacks:
    def __init__(self, threshold):
        self.stacks = []
        self.threshold = threshold
        self.size = []
        self.numOfStacks = 0
    
    def push(self, value):
        if (self.size % self.threshold == 0):
            self.stacks.append(StackByList())
            self.numOfStacks += 1
        
        self.stacks[self.numOfStacks-1].push(value)
        self.size += 1
        
    def pop(self):
        if (self.size == 0):
            return 'the stack is empty'
        
        self.size -= 1
        return self.stacks[self.numOfStacks-1].pop()
    
    def popAt(self, index):
        if (self.size == 0):
            return 'the stack is empty'
        if (self.numOfStacks < index):
            return 'specific stack does not exist'
        
        self.size -= 1
        return self.stacks[index].pop()
    
    def peek(self):
        if (self.size == 0):
            return 'the stack is empty'
        
        return self.stacks[self.numOfStacks-1].peek()
    
    def printList(self):
        for stack in self.stacks:
            stack.printList()

a3 = SetOfStacks(4)
a3.printList()

a3.push(1)
a3.push(2)
a3.push(3)
a3.push(4)
a3.push(5)
a3.push(6)
a3.push(7)
a3.printList()
print('')

a3.pop()
a3.printList()
print('')

a3.popAt(0)
a3.printList()
print('')

a3.pop()
a3.printList()
print('')

a3.pop()
a3.printList()
print('')

a3.pop()
a3.printList()
print('')
```
