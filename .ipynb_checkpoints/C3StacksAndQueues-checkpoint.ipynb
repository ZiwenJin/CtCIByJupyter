## 3.0

```python
class StackByList:
    def __init__(self):
        self.values = []
        self.size = 0
        
    def push(self, value):
        self.values.append(value)
        self.size += 1
        
    def pop(self):
        if (self.size == 0):
            print('this stack is empty 1')
            return
        
        self.size -= 1
        return self.values.pop()
    
    def peek(self):
        if (self.size == 0):
            print('this stack is empty 2')
            return
        
        return self.values[self.size-1]
    
    def isEmpty(self):
        return self.size == 0

    def printList(self):
        print(self.values)
```

## 3.1 Three in One 2

```python
class MultiStack:      
    def __init__(self, numberOfStacks, defaultSize):
        self.numberOfStacks = numberOfStacks
        self.values = [0] * numberOfStacks * defaultSize
        self.start = []
        self.capacity = []
        self.size = []
        for i in range(numberOfStacks):
            self.start.append(defaultSize * i)
            self.capacity.append(defaultSize)
            self.size.append(0)
    
    def push(self, stackNum, value):
        if (self.allStacksAreFull()):
            print('all stacks are full')
            return None
        
        if (self.isFull(stackNum)):
            self.expand(stackNum)
            
        self.size[stackNum] += 1
        self.values[self.lastElementIndex(stackNum)] = value
    
    def pop(self, stackNum):
        if (self.isEmpty(stackNum)):
            print('this stack is empty')
            return None
        
        value = self.values[self.lastElementIndex(stackNum)]
        self.values[self.lastElementIndex(stackNum)] = 0
        self.size[stackNum] -= 1
        return value
    
    def peek(self, stackNum):
        return self.values[self.lastElementIndex(stackNum)]
    
    def shift(self, stackNum):
        print('   Shifting ', stackNum)
        
        if (self.size[stackNum] >= self.capacity[stackNum]):
            self.shift((stackNum + 1) % self.numberOfStacks)
            self.capacity[stackNum] += 1
        
        index = self.lastCapacityIndex(stackNum)
        while (self.isWithinStackCapacity(stackNum, index)):
            self.values[index] = self.values[self.previousIndex(index)]
            index = self.previousIndex(index)
        
        self.values[self.start[stackNum]] = 0
        self.start[stackNum] = self.nextIndex(self.start[stackNum])
        self.capacity[stackNum] -= 1
        
    def expand(self, stackNum):
        self.shift((stackNum + 1) % self.numberOfStacks)
        self.capacity[stackNum] += 1
    
    def isWithinStackCapacity(self, stackNum, index):
        if (index < 0 or index >= len(self.values)):
            return False

        contiguousIndex = index + len(self.values) if (index < self.start[stackNum]) else index
        end = self.start[stackNum] + self.capacity[stackNum]
        return self.start[stackNum] <= contiguousIndex and contiguousIndex < end

    def lastCapacityIndex(self, stackNum):
        return self.adjustIndex(self.start[stackNum] + self.capacity[stackNum] - 1)

    def lastElementIndex(self, stackNum):
        return self.adjustIndex(self.start[stackNum] + self.size[stackNum] - 1)

    def isFull(self, stackNum):
        return self.size[stackNum] == self.capacity[stackNum]

    def isEmpty(self, stackNum):
        return self.size[stackNum] == 0
    
    def allStacksAreFull(self):
        return sum(self.size) == len(self.values)
    
    def adjustIndex(self, index):
        max = len(self.values)
        return ((index % max) + max) % max
    
    def nextIndex(self, index):
        return self.adjustIndex(index + 1)
    
    def previousIndex(self, index):
        return self.adjustIndex(index - 1)
    
    def printList(self):
        print(self.values)
    
'''
a = MultiStack(3, 5)
a.printList()

a.push(2, 5)
a.push(2, 6)
a.push(2, 7)
a.push(2, 8)
a.push(2, 9)
a.printList()

a.push(2, 10)
a.printList()

a.push(2, 11)
a.printList()

print(a.pop(2))
a.printList()

a.push(2, 11)
a.printList()

a.push(0,1)
a.printList()

a.push(0,2)
a.push(0,3)
a.push(0,4)
a.push(0,5)
a.push(0,6)
a.push(0,7)
a.printList()

a.push(1,-1)
a.printList()

a.pop(1)
a.printList()
'''
```

## 3.2 Stack Min

```python
# use a stack to store mins
# or store the min beneath it for each element

class StackByListWithMin:
    def __init__(self):
        self.values = []
        self.mins = StackByList()
        self.size = 0
        
    def push(self, value):
        self.values.append(value)
        self.size += 1
        if (self.size == 1):
            self.mins.push(value)
        elif (value < self.mins.peek()):
            self.mins.push(value)
        
    def pop(self):
        if (self.size == 0):
            print('this stack is empty 3')
            return
        
        value = self.values.pop()
        self.size -= 1
        if (value == self.mins.peek()):
            self.mins.pop()
        return value
    
    def peek(self):
        if (self.size == 0):
            print('this stack is empty 4')
            return
        
        return self.values[self.size-1]
    
    def min(self):
        if (self.size == 0):
            print('this stack is empty 5')
            return
        
        return self.mins.peek()
    
    def printList(self):
        print(self.values)

'''
a = StackByListWithMin()

a.push(4)
a.printList()
print(a.min())
print('')

a.push(3)
a.printList()
print(a.min())
print('')

a.push(5)
a.printList()
print(a.min())
print('')

a.push(2)
a.printList()
print(a.min())
print('')

a.push(4)
a.printList()
print(a.min())
print('')

print(a.pop())
a.printList()
print(a.min())
print('')

print(a.pop())
a.printList()
print(a.min())
print('')

print(a.pop())
a.printList()
print(a.min())
print('')

print(a.pop())
a.printList()
print(a.min())
print('')

print(a.pop())
a.printList()
print(a.min())
print('')
'''
```

## 3.3 Stack of Plates

```python
class SetOfStacks:
    def __init__(self, threshold):
        self.stacks = []
        self.stacks.append(StackByList())
        self.threshold = threshold
        self.size = 0
        self.numOfStacks = 1
    
    def push(self, value):
        if (self.size > 0 and self.size % self.threshold == 0):
            self.stacks.append(StackByList())
            self.size = 0
            self.numOfStacks += 1
        
        self.stacks[self.numOfStacks-1].push(value)
        self.size += 1
        
    def pop(self):
        if (self.numOfStacks == 1 and self.size == 0):
            return 'the stack is empty'
        
        if self.size == 0:
            self.numOfStacks -= 1
            self.size = self.threshold
        self.size -= 1
        return self.stacks[self.numOfStacks-1].pop()
    
#     def popAt(self, index):
#         if (self.size == 0):
#             return 'the stack is empty'
#         if (self.numOfStacks < index):
#             return 'specific stack does not exist'
        
#         self.size -= 1
#         return self.stacks[index].pop()
    
    def peek(self):
        if (self.numOfStacks == 1 and self.size == 0):
            return 'the stack is empty'
        
        return self.stacks[self.numOfStacks-1].peek() if self.size > 0 else self.stacks[self.numOfStacks-2].peek()
    
    def printList(self):
        for stack in self.stacks:
            stack.printList()

a3 = SetOfStacks(4)
a3.printList()
print('')

a3.push(1)
a3.push(2)
a3.push(3)
a3.push(4)
a3.push(5)
a3.push(6)
a3.push(7)
a3.printList()
print('')

a3.pop()
a3.printList()
print('')

# a3.popAt(0)
# a3.printList()
# print('')

a3.pop()
a3.printList()
print('')

a3.pop()
a3.printList()
print('')

print(a3.peek())

print(a3.pop())
a3.printList()
print('')
```

# 3.4 Queue via Stacks

```python
class MyQueue:
    def __init__(self):
        self.stackNorm = StackByList()
        self.stackReverse = StackByList()
        self.start = None
        self.size = 0
    
    def add(self, value):
        if (self.size == 0):
            self.start = value
        self.size += 1
        self.stackNorm.push(value)
    
    def remove(self):
        if (self.size == 0):
            return 'this queue is empty'
            
        while (not self.stackNorm.isEmpty()):
            self.stackReverse.push(self.stackNorm.pop())
        
        self.size -= 1
        value = self.stackReverse.pop()
        self.start = self.stackReverse.peek()
        
        while (not self.stackReverse.isEmpty()):
            self.stackNorm.push(self.stackReverse.pop())
        
        return value
    
    def peek(self):
        return self.start
    
    def printList(self):
        print(self.stackNorm.values)
    
a4 = MyQueue()
a4.add(1)
a4.add(2)
a4.add(3)
a4.printList()

print(a4.remove())
a4.printList()

a4.add(4)
a4.printList()

print(a4.peek())
```

## 3.4 Queue via Stacks 2

```python
# let elements stay in stack2
class MyQueue2:
    def __init__(self):
        self.stackNew = StackByList()
        self.stackOld = StackByList()
    
    def size(self):
        return self.stackNew.size + self.stackOld.size

    def add(self, value):
        self.stackNew.push(value)
        
    def shiftStack(self):
        if (self.stackOld.isEmpty()):
            while (not self.stackNew.isEmpty()):
                self.stackOld.push(self.stackNew.pop())
    
    def remove(self):
        self.shiftStack()
        return self.stackOld.pop()
    
    def peek(self):
        self.shiftStack()
        return self.stackOld.peek()

    def isEmpty(self):
        return self.size() == 0
    
    def printList(self):
        s1 = copy.deepcopy(self.stackNew)
        s2 = copy.deepcopy(self.stackOld)
        
        result = ''
        while (not s2.isEmpty()):
            result += str(s2.pop()) + ', '
        
        while (not s1.isEmpty()):
            s2.push(s1.pop())
            
        while (not s2.isEmpty()):
            result += str(s2.pop()) + ', '
        
        print(result)
    
a42 = MyQueue2()

a42.add(1)
a42.add(2)
a42.add(3)
a42.printList()
print('')

print(a42.peek())
print(a42.remove())
a42.printList()
print('')

print(a42.peek())
print(a42.remove())
a42.printList()
print('')

print(a42.peek())
print(a42.remove())
a42.printList()
print('')

print(a42.peek())
print(a42.remove())
a42.printList()
print('')
```

# 3.5 Sort Stack

```python
def sortStack(stack):
    stackSmallest = StackByList()
    
    while (not stack.isEmpty()):
        curValue = stack.pop()
        
        while ((not stackSmallest.isEmpty()) and (stackSmallest.peek() > curValue)):
            stack.push(stackSmallest.pop())
        stackSmallest.push(curValue)
        
    while (not stackSmallest.isEmpty()):
        stack.push(stackSmallest.pop())
        
    return stack

a5 = StackByList()
a5.push(3)
a5.push(2)
a5.push(5)
a5.push(1)
a5.push(4)
a5.printList()

sortStack(a5)
print(a5.pop(), end=', ')
print(a5.pop(), end=', ')
print(a5.pop(), end=', ')
print(a5.pop(), end=', ')
print(a5.pop(), end=', ')
```

# 3.6 Animal Shelter

```python
class AnimalShelter:
    class Animal:
        def __init__(self, animalType, index):
            self.animalType = animalType
            self.order = index
        
    def __init__(self):
        self.index = 0
        self.dogs = MyQueue2()
        self.cats = MyQueue2()
    
    def enqueue(self, animalType):
        self.index += 1
        if (animalType == 'dog'):
            self.dogs.add(self.Animal(animalType, self.index))
        elif(animalType == 'cat'):
            self.cats.add(self.Animal(animalType, self.index))
        else:
            return 'invalid animal type'
    
    def dequeueDog(self):
        return self.dogs.remove()
    
    def dequeueCat(self):
        return self.cats.remove()
    
    def dequeueAny(self):
        if (self.dogs.size() == 0 and self.cats.size() == 0):
            raise ValueError('no animals')
        elif (self.dogs.size() == 0):
            return self.dequeueCat()
        elif (self.cats.size() == 0):
            return self.dequeueDog()
        
        dogOldest = self.dogs.peek().order
        catOldest = self.cats.peek().order
        
        if (dogOldest <= catOldest):
            return self.dequeueDog()
        else:
            return self.dequeueCat()
    
    def printList(self):
        d = copy.deepcopy(self.dogs)
        c = copy.deepcopy(self.cats)
        
        print('Dogs: ', end='')
        while (not (d.size() == 0)):
            print('[', d.peek().animalType, ',', str(d.peek().order), ']', ',', end='')
            d.remove()
        print('')
        
        print('Cats: ', end='')
        while (not (c.size() == 0)):
            print('[', c.peek().animalType, ',', str(c.peek().order), ']', ',', end='')
            c.remove()
        print('')


        
a6 = AnimalShelter()
a6.printList()
print('')

a6.enqueue('dog')
a6.enqueue('dog')
a6.enqueue('cat')
a6.enqueue('dog')
a6.enqueue('cat')
a6.enqueue('dog')
a6.printList()
print('')

print(a6.dequeueAny().animalType)
a6.printList()
print('')

print(a6.dequeueCat().animalType)
a6.printList()
print('')

print(a6.dequeueDog().animalType)
a6.printList()
print('')

print(a6.dequeueAny().animalType)
a6.printList()
print('')

print(a6.dequeueAny().animalType)
a6.printList()
print('')

print(a6.dequeueAny().animalType)
a6.printList()
print('')

print(a6.dequeueAny().animalType)
a6.printList()
print('')
```
